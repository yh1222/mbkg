```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
base_dir = Path('D:\projects\MGKG\code\Train')
base_dir
```

```python
results_path = [f for f in base_dir.rglob('*.csv') if Path(f).is_file()][3:]
results_path
```

```python
# Loss comparison
TransE_l2 = pd.read_csv(results_path[0], sep=' ', header=None)[4].values.tolist()
TransR = pd.read_csv(results_path[4], sep=' ', header=None)[4].values.tolist()
legend = ['TransE_l2', 'TransR']
plt.plot(TransE_l2)
plt.plot(TransR)
plt.title('Running Loss')
plt.xlabel('1000 Steps')
plt.ylabel('Loss')
plt.legend(legend, loc='best')
plt.savefig(base_dir / 'Loss_compare.png')
plt.show()
```

```python
# Neg_loss comparison
TransE_l2 = pd.read_csv(results_path[1], sep=' ', header=None)[4].values.tolist()
TransR = pd.read_csv(results_path[5], sep=' ', header=None)[4].values.tolist()
legend = ['TransE_l2', 'TransR']
plt.plot(TransE_l2)
plt.plot(TransR)
plt.title('Running Neg_loss')
plt.xlabel('1000 Steps')
plt.ylabel('Neg_loss')
plt.legend(legend, loc='best')
plt.savefig(base_dir / 'Neg_loss_compare.png')
plt.show()
```

```python
# Pos_loss comparison
TransE_l2 = pd.read_csv(results_path[2], sep=' ', header=None)[4].values.tolist()
TransR = pd.read_csv(results_path[6], sep=' ', header=None)[4].values.tolist()
legend = ['TransE_l2', 'TransR']
plt.plot(TransE_l2)
plt.plot(TransR)
plt.title('Running Pos_loss')
plt.xlabel('1000 Steps')
plt.ylabel('Pos_loss')
plt.legend(legend, loc='best')
plt.savefig(base_dir / 'Pos_loss_compare.png')
plt.show()
```

```python
# Regularization comparison
TransE_l2 = pd.read_csv(results_path[3], sep=' ', header=None)[4].values.tolist()
TransR = pd.read_csv(results_path[7], sep=' ', header=None)[4].values.tolist()
legend = ['TransE_l2', 'TransR']
plt.plot(TransE_l2)
plt.plot(TransR)
plt.title('Running Regularization')
plt.xlabel('1000 Steps')
plt.ylabel('Regularization')
plt.legend(legend, loc='best')
plt.savefig(base_dir / 'Regularization_compare.png')
plt.show()
```

```python
# TransE_l2 Loss
TransE_l2_Loss = pd.read_csv(results_path[0], sep=' ', header=None)[4].values.tolist()
TransE_l2_Neg_loss = pd.read_csv(results_path[1], sep=' ', header=None)[4].values.tolist()
TransE_l2_Pos_loss = pd.read_csv(results_path[2], sep=' ', header=None)[4].values.tolist()
legend = ['Loss', 'Neg_loss', 'Pos_loss']
plt.plot(TransE_l2_Loss)
plt.plot(TransE_l2_Neg_loss)
plt.plot(TransE_l2_Pos_loss)
plt.title('TransE_l2')
plt.xlabel('1000 Steps')
plt.ylabel('Vary Loss')
plt.legend(legend, loc='best')
plt.savefig(base_dir / 'TransE_l2_Loss.png')
plt.show()
```
